```python
import pandas as pd
import numpy as np

# Assign data
data = {
    'Name': ['Sai', 'Yuvasri', 'Poovannan', 'Sabharish', 'Sangamithra', 'Rithika', 'Rubasri'],
    'Age': [20, 18, 19, 20, 18, 17, 19],
    'Gender': ['M', 'F', 'M', 'M', 'F', 'F', 'F'],
    'Marks': [90, 76, np.nan, 74, 65, np.nan, 71]  # Using np.nan is better practice
}

# Convert into dataframe
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Calculate average (The Pandas Way)
# This handles the math automatically and ignores missing values
avg_marks = df['Marks'].mean()

# Replace missing values
df['Marks'] = df['Marks'].fillna(avg_marks)

# Display data
print("\nReplacing NaN with average marks:\n", df)
```

```python
#Filter top scoring students
df=df[df['Marks']>=75]

#Remove age row
df=df.drop(['Age'],axis=1)

#Display data
print("\nFiltering and Dropping func:\n",df)
```
