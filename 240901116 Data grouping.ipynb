```python
import pandas as pd

df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # numeric values
})

# Group by Animal and sum numeric columns
b = df.groupby(['Animal']).sum()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using sum\n", b)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # numeric values
})

# Group by Animal and count non-NaN values in each column
c = df.groupby(['Animal']).count()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using count\n", c)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # numeric values
})

# Group by Animal and get the first row in each group
d = df.groupby(['Animal']).first()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using first\n", d)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # numeric values
})

# Group by Animal and get the last row in each group
e = df.groupby(['Animal']).last()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using last\n", e)
```
