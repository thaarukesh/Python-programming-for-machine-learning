```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    [[1, 2, 3], [4, 5, 6], [7, 8, 9], [np.nan, np.nan, np.nan]],
    columns=['A', 'B', 'C']
)

print("Original DataFrame\n", df)
print("\nAggregate function: aggregate()\n", df.aggregate(['sum']))
print("\nAggregate function: alias agg()\n", df.agg(['sum']))
```

```python
# Aggregate function to produce sum, min and max
a = df.agg(['sum', 'min', 'max'])

print("\nOriginal DataFrame\n", df)
print("\nAggregate function to produce sum, min and max\n", a)
```

```python
# Different aggregation per column
b = df.agg({
    'A': ['sum', 'min'],
    'B': ['min', 'max']
})

print("\nOriginal DataFrame\n", df)
print("\nDifferent aggregations per column\n", b)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    [[1, 2, 3], [4, 5, 6], [7, 8, 9], [np.nan, np.nan, np.nan]],
    columns=['A', 'B', 'C']
)

# Apply different functions per column
c = df.agg({
    'A': max,
    'B': 'min',
    'C': np.mean
})

# Rename the resulting index
c.index = ['x', 'y', 'z']

print("\nOriginal DataFrame\n", df)
print("\nAggregate different functions over the columns with renamed index\n", c)
```

```python
d = df.agg("sum", axis="columns")

print("\nOriginal DataFrame\n", df)
print("\nAggregate - Sum (row wise)\n", d)
```

```python
# Column-wise sum
e = df.agg("sum", axis=0)  # axis=0 is column-wise

print("\nOriginal DataFrame\n", df)
print("\nAggregate - Sum (column wise)\n", e)
```

```python
print("\nOriginal DataFrame\n", df)
print("\nSummary of aggregate functions using describe()\n", df.describe())
```
